```python
# Question 2.12

import math as m

x = float(input("Input a number for x: "))
y = float(input("Input a number for y: ")) # user input for number x and y

angle = m.degrees(m.atan2(y, x)) # calculation for angle

if angle < 0:
    angle = 360 + angle # converting to 0 to 360 degree range

if 0 <= angle < 90:
    print(f"Your {angle} degree angle is in the first quadrant")
elif  90 <= angle < 180:
    print(f"Your {angle} degree angle is in the second quadrant")
elif  180 <= angle < 270:
    print(f"Your {angle} degree angle is in the third quadrant")
elif  270 < angle < 360:
    print(f"Your {angle} degree angle is in the fourth quadrant") 
else:
    print("error") # assigning quadrant and printing result
```

```python
X = [1.36,  1.46,  1.41,  2.12,  0.42,  0.98,  0.45,  1.97, -0.24, 2.18]

X.append(0.51)
X.remove(2.12)

print(f"The value of index number 6 in the list is {X[6]}")

for i in range(len(X)):
    if (X[i] == 1.46):
        print(f"The value of 1.46 has the index number {i}")

A = slice(2, 8)
total_sum = 0

for i in range(len(X[A])):
    if (X[A][i] * 100 % 2 == 0):
        total_sum += X[A][i]

print(f"The total sum is {total_sum}")
```

```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

set_night1 = set(night1)
set_night2 = set(night2)
combined_set = set_night1.intersection(set_night2)

if 5 in combined_set:
    print("5 is in the combined set.")
else:
    print("Nope, 5 is not in the combined set.")

if 5 in set_night1:
    print("5 is in the first set.")
else:
    print("5 is NOT in the first set")
```

```python
fibonacci = [1, 2]

fibonacci_calculated = 0

n_loop = 98

while n_loop > 0:
    fibonacci_calculated += fibonacci[-2]
    fibonacci_calculated += fibonacci[-1]
    fibonacci.append(fibonacci_calculated)
    fibonacci_calculated = 0
    n_loop -= 1

print(f"The 100th number in the fibonacci sequence is {fibonacci[len(fibonacci)-1]}")
```
